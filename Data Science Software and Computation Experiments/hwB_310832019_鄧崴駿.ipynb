```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
ax = plt.gca()
ax.set_facecolor('black')

t = np.linspace(0, 2*np.pi, 50)
plt.plot(np.cos(t), np.sin(t), color='#86ddeb', linestyle='--', marker="x", linewidth=3)
plt.plot(.2*np.cos(t), .2*np.sin(t) + 0.30901699, color='#86ddeb', linewidth=5)
plt.plot(.35*np.cos(t), .35*np.sin(t) + 0.30901699, color='#d2f2f9', linewidth=3)
plt.plot(.4*np.cos(t), .4*np.sin(t) + 0.30901699, color='#86ddeb')
plt.plot(.5*np.cos(t), .5*np.sin(t) + 0.30901699, color='#d2f2f9')
plt.plot(.55*np.cos(t), .55*np.sin(t) + 0.30901699, color='#86ddeb', linewidth=3)
plt.plot(1.3*np.cos(t), 1.3*np.sin(t), color='#86ddeb')
plt.plot(1.35*np.cos(t), 1.35*np.sin(t), color='#92e9f7', marker="X")

for i in range(3):
    t0 = 0.1*np.pi - 6*np.pi/10*i
    x0,y0 = 1.15*np.cos(t0), 1.15*np.sin(t0)
    plt.plot(.15*np.cos(t)+x0, .15*np.sin(t)+y0, '#d2f2f9')
    plt.plot(.1*np.cos(t)+x0, .1*np.sin(t)+y0, '#d2f2f9', linewidth=3)
    
    t1 = np.linspace(0.5*np.pi, 2.5*np.pi, 4)
    plt.plot(0.1*np.cos(t1)+x0, 0.1*np.sin(t1)+y0, color='#d2f2f9', linewidth=2)


t = np.linspace(0, 4*np.pi, 6)
x=np.cos(t+np.pi/10)
y=np.sin(t+np.pi/10)
y[3] = 0.30901699
plt.plot(x, y, color='#d2f2f9', linewidth=5)

plt.plot([0,0], [0.30901699, -1], color='#d2f2f9', linewidth=5)

plt.plot([0, 0.08, 0, -0.08, 0], [1.3, 1.1, 0.9, 1.1, 1.3], color='#d2f2f9', linewidth=3)
```
